```python
# ==================================
# Regular expression example
# ==================================
```

```python
import re
```

```python
texto = "UE=12 c-rnti=0x4601 RSSI=-73 dBm"
```

```python
m = re.search(r"UE=(\d+).*c-rnti=(0x[0-9a-fA-F]+)", texto)
if m:
    print("UE:", m.group(1))
    print("C-RNTI:", m.group(2))
```

```python
texto = '''
Puedes contactar conmigo en los siguientes correos:

- personal: alvaro.duran@example.com
- trabajo: a.duran+dev@company-tech.io
- alternativo: user_test123@mail.co.uk
'''
```

```python
p = re.compile(r"([\w\.+]+)@([a-zA-Z_-]+\.[a-zA-Z\.-]+)")
```

```python
p
```

```python
m = re.search(p, texto)
if m:
    print("Nombre:", m.group(1))
    print("Dominio:", m.group(2))
```

```python
m = re.search(p, texto)
if m:
    print("Nombre, Dominio:", m.groups())
```

```python
m = re.search(p, texto)
if m:
    print(f"Coincidencia: {m.group(0)}")
    print(f"Nombre: {m.group(1)}")
    print(f"Dominio: {m.group(1)}")
```

https://docs.python.org/3.14/library/re.html#functions

```python
# ==================================
# findall:
# - Return all non-overlapping matches of pattern in string, as a list of strings or tuples
# ==================================
```

```python
m = re.findall(p, texto, flags=0)
```

```python
for res in m:
    print(res)
```

```python
# ==================================
# finditer
# - Return an iterator yielding Match objects over all non-overlapping matches for the RE pattern in string
# ==================================
```

```python
m = re.finditer(p, texto, flags=0)
```

```python
for i, res in enumerate(m):
    print(f"# {i}")
    print(f"Coincidencia: {res.group(0)}")
    print(f"Nombre: {res.group(1)}")
    print(f"Dominio: {res.group(1)}")
```
