```python
import time
import asyncio
```

```python
# Synchronous vs asynchronous example
```

```python
# ==================================
# (1) Synchronous version
# ==================================
```

```python
def task_sync(name: str, delay: float) -> None:
    print(f"- Task {name} started (it will wait {delay} s)")
    time.sleep(delay) # Wait for some external task
    print(f"- Task {name} finished after {delay} s")

def run_sync() -> None:
    print("---------------------")
    print("Synchronous execution")
    print("---------------------\n")
    
    t0 = time.perf_counter()

    task_sync("A", 5)
    task_sync("B", 10)
    task_sync("C", 7)

    t1 = time.perf_counter()
    
    print(f"\nTotal synchronous time: {t1 - t0:.2f} s")
```

```python
run_sync()
```

```python
# ==================================
# (2.1) Asynchronous (asyncio) version (with gather)
# ==================================
```

```python
async def task_async(name: str, delay: float) -> None:
    print(f"- Task {name} started (it will wait {delay} s)")
    await asyncio.sleep(delay) # Wait for some external task
    print(f"- Task {name} finished after {delay} s")

async def run_async() -> None:
    print("---------------------")    
    print("asyncio execution (with gather)")
    print("---------------------\n")
    
    t0 = time.perf_counter()

    await asyncio.gather(
        task_async("A", 5),
        task_async("B", 10),
        task_async("C", 7),
    )

    t1 = time.perf_counter()
    print(f"\nTotal asyncio time: {t1 - t0:.2f} s")
```

```python
await run_async()
```

```python
# ==================================
# (2.2) Asynchronous (asyncio) version (with TaskGroup)
# ==================================
```

```python
async def task_async(name: str, delay: float) -> None:
    print(f"- Task {name} started (it will wait {delay} s)")
    await asyncio.sleep(delay) # Wait for some external task
    print(f"- Task {name} finished after {delay} s")

async def run_async() -> None:
    print("---------------------")    
    print("asyncio execution (with TaskGroup)")
    print("---------------------\n")
    
    t0 = time.perf_counter()

    async with asyncio.TaskGroup() as tg:
        tg.create_task(task_async("A", 5))
        tg.create_task(task_async("B", 10))
        tg.create_task(task_async("C", 7))
        
    t1 = time.perf_counter()
    print(f"\nTotal asyncio time: {t1 - t0:.2f} s")
```

```python
await run_async()
```

```python
# ==================================
# (2.3) Asynchronous (asyncio) version (with create_task)
# ==================================
```

```python
async def task_async(name: str, delay: float) -> None:
    print(f"- Task {name} started (it will wait {delay} s)")
    await asyncio.sleep(delay) # Wait for some external task
    print(f"- Task {name} finished after {delay} s")

async def run_async() -> None:
    print("---------------------")    
    print("asyncio execution (with create_task)")
    print("---------------------\n")
    
    t0 = time.perf_counter()

    # Create tasks (they start immediately)
    task1 = asyncio.create_task(task_async("A", 5))
    task2 = asyncio.create_task(task_async("B", 10))
    task3 = asyncio.create_task(task_async("C", 7))

    # Wait for them (order here does NOT matter for concurrency)
    await task1
    await task2
    await task3    
        
    t1 = time.perf_counter()
    print(f"\nTotal asyncio time: {t1 - t0:.2f} s")
```

```python
await run_async()
```
