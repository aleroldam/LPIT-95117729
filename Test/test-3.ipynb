```python
# ==================================
# Functions
# ==================================
```

```python
# - Multiple return values 
# - Default parameter value
def calcular(a, b=5):
    return a + b, a * b
```

```python
# Positional arguments
suma, producto = calcular(5, 10)
```

```python
print(suma, producto)
```

```python
# Named arguments
suma, producto = calcular(a=5, b=10)
```

```python
print(suma, producto)
```

```python
# Default parameter value
suma, producto = calcular(a=2)
```

```python
print(suma, producto)
```

```python
# Default parameter value
suma, producto = calcular(2)
```

```python
print(suma, producto)
```

```python
# Arbitrary number of positional arguments: *args
def promedio(*notas):
    return sum(notas) / len(notas)
```

```python
print(promedio(5, 7, 9))
```

```python
# Arbitrary number of keyword arguments: **kwargs
def promedio(y, m, *notas, **nombres) -> float:
    
    print("y =", y)
    print("m =", m)
    print("Notas =", notas)
    print("Nombres =", nombres)

    return sum(notas) / len(notas)
```

```python
promedio(2026, 3, 8.1, 5.6, Nombre1="Dean", Nombre2="Dave", Nombre3="Chuck")
```

```python
# ==================================
# Lambda functions
# ==================================
```

```python
# Basic example
```

```python
square = lambda x: x * x
```

```python
print(square(5))
```

```python
# Anonymous function inside another function
```

```python
def f_multiplier(n):
  return lambda a : a * n
```

```python
f_doubler = f_multiplier(2)
```

```python
print(f_doubler(4.25))
```

```python
f_doubler(10.2)
```
