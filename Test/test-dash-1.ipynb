```python
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import pandas as pd
import polars as pl
```

```python
# ==================================
# Plot data using Pandas
# ==================================
```

```python
# Sample dataframe
```

https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv

```python
dfa = pd.read_csv('gapminder_unfiltered.csv')
```

```python
dfa
```

```python
dfa.country.unique()
```

```python
dfa[dfa.country=='Zimbabwe']
```

```python
app = Dash()

# Requires Dash 2.17.0 or later
app.layout = [
    html.H1(children='Dash app with data (Pandas)', style={'textAlign':'center'}),
    dcc.Dropdown(dfa.country.unique(), 'Canada', id='dropdown-selection'),
    dcc.Graph(id='graph-content')
]

@callback(
    Output('graph-content', 'figure'),
    Input('dropdown-selection', 'value')
)
def update_graph(value):
    dff = dfa[dfa.country==value]
    return px.line(dff, x='year', y='pop')

# Note: if app.run yields an error try changing the server port

if __name__ == '__main__':
    app.run(host="127.0.0.1", port=8051, debug=False)
```

```python
# Sample dataframe
```

https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv

```python
dfo = pl.read_csv('gapminder_unfiltered.csv')
```

```python
dfo
```

```python
dfo.unique(subset="country")
```

```python
dfo.filter(pl.col("country")=='Zimbabwe')
```

```python
app = Dash()

# Requires Dash 2.17.0 or later
app.layout = [
    html.H1(children='Dash app with data (Polars)', style={'textAlign':'center'}),
    dcc.Dropdown(dfo["country"].unique().to_list(), 'Canada', id='dropdown-selection'),
    dcc.Graph(id='graph-content')
]

@callback(
    Output('graph-content', 'figure'),
    Input('dropdown-selection', 'value')
)
def update_graph(value):
    dff = dfo.filter(pl.col("country")==value)
    return px.line(dff, x='year', y='pop')

# Note: if app.run yields an error try changing the server port

if __name__ == '__main__':
    app.run(host="127.0.0.1", port=8052, debug=False)
```

```python
# --------------------------------
# Visualize data in a table using Pandas
# --------------------------------
```

https://raw.githubusercontent.com/plotly/datasets/master/gapminder2007.csv

```python
# Import packages
from dash import Dash, html
import dash_ag_grid as dag
import pandas as pd

# Incorporate data
dfn = pd.read_csv('gapminder2007.csv')

# Initialize the app
app = Dash()

# App layout
app.layout = [
    html.Div(children='Dash app with data (Pandas)'),
    dag.AgGrid(
        rowData=dfn.to_dict('records'),
        columnDefs=[{"field": i} for i in dfn.columns]
    )
]

# Note: if app.run yields an error try changing the server port

if __name__ == '__main__':
    app.run(host="127.0.0.1", port=8054, debug=False)
```

https://raw.githubusercontent.com/plotly/datasets/master/gapminder2007.csv

```python
# Import packages
from dash import Dash, html
import dash_ag_grid as dag
import polars as pl

# Incorporate data
dfl = pl.read_csv('gapminder2007.csv')

# Initialize the app
app = Dash()

# App layout
app.layout = [
    html.Div(children='Dash app with data (Polars)'),
    dag.AgGrid(
        rowData=dfl.to_dicts(),
        columnDefs=[{"field": i} for i in dfl.columns]
    )
]

# Note: if app.run yields an error try changing the server port

if __name__ == '__main__':
    app.run(host="127.0.0.1", port=8059, debug=False)
```
